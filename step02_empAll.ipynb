```python
import numpy as np
import pandas as pd
```

```python
# 외부에 있는 파일 불러와서 DataFrame 생성

df = pd.read_csv("../00.dataSet/empAll.csv")
df
```

```python
# 연봉 column 추가(COMM 포함)

df['ANNUAL'] = np.where(df['COMM'] > 0, df['SAL']*12+df['COMM'], df['SAL']*12)
df
```

```python
# HIREDATE를 yyyy-mm-dd 형태로 변경

H_DATE = []
for v in df['HIREDATE']:
    d = v.split('-')
    H_DATE.append(d[2]+'-'+d[0]+'-'+d[1])
    
df['HIREDATE'] = pd.Series(H_DATE)
df
```

```python
# job 종류 추출

dfg = pd.DataFrame({'COUNT' : df.groupby('JOB').size() }).reset_index()
dfg
```
